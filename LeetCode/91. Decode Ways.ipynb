# 91. Decode Ways

A message containing letters from A-Z is being encoded to numbers using the following mapping:

'A' -> 1
'B' -> 2
...
'Z' -> 26
Given an encoded message containing digits, determine the total number of ways to decode it.

For example,
Given encoded message "12", it could be decoded as "AB" (1 2) or "L" (12).

The number of ways decoding "12" is 2.



## Solution. Dynamic Programming with Time Complexity of O(n)

```python
class Solution(object):
    def numDecodings(self, s):
        """
        :type s: str
        :rtype: int
        """
        if s == "":
            return 0
        
        length = len(s)
        result = [0 for i in range(length)]
        
        if s[0] == '0':
            return 0
        else:
            result[0] = 1
        
        for i in range(1, length):
            # calculate number of encoding ending at ith element
            if s[i-1] == '0':
                if i == 1:
                    return 0
                else:
                    if s[i] == '0':
                        return 0
                    else:
                        result[i] = result[i-1] 
                        
            elif s[i-1] == '1':
                if s[i] == '0':
                    if i == 1: 
                        result[i] = 1
                    else:
                        result[i] = result[i-2]
                else:
                    if i == 1:
                        result[i] = 2
                    else:    
                        result[i] = result[i-2] + result[i-1]
                        
            elif s[i-1] == '2':
                if s[i] in '123456':
                    if i == 1:
                        result[i] = 2
                    else:
                        result[i] = result[i-2] + result[i-1]
                elif s[i] == '0':
                    if i == 1:
                        result[i]  = 1
                    else: 
                        result[i] = result[i-2]
                elif s[i] in '789':
                    if i == 1:
                        result[i] = 1
                    else:
                        result[i] = result[i-1]
                        
            elif s[i-1] in '3456789':
                if s[i] == '0':
                    return 0
                else:
                    if i == 1:
                        result[i] = 1
                    else:
                        result[i] = result[i-1]
        
        return result[length-1]
```

## Solution 2. Minor Enhancement

```python
class Solution(object):
    def numDecodings(self, s):
        """
        :type s: str
        :rtype: int
        """
        if s == "":
            return 0
        
        length = len(s)
        result = [0 for i in range(length)]
        
        if s[0] == '0':
            return 0
        else:
            result[0] = 1
            
            if len(s) == 1: 
                return 1
            elif s[0] == '1':
                if s[1] == '0':
                    result[1] = 1
                else:
                    result[1] = 2
            elif s[0] == '2':
                if s[1] in '0789':
                    result[1] = 1
                elif s[1] in '123456':
                    result[1] = 2
            elif s[0] in '3456789':
                if s[1] == '0':
                    return 0
                else:
                    result[1] = 1
   
        for i in range(2, length):
            # calculate number of encoding ending at ith element
            if s[i-1] == '0':
                if s[i] == '0':
                    return 0
                else:
                    result[i] = result[i-1]    
            elif s[i-1] == '1':
                if s[i] == '0':
                    result[i] = result[i-2]
                else:   
                    result[i] = result[i-2] + result[i-1]
                        
            elif s[i-1] == '2':
                if s[i] in '123456':
                    result[i] = result[i-2] + result[i-1]
                elif s[i] == '0':
                    result[i] = result[i-2]
                elif s[i] in '789':
                    result[i] = result[i-1]
                        
            elif s[i-1] in '3456789':
                if s[i] == '0':
                    return 0
                else:
                    result[i] = result[i-1]
    
        return result[length-1]
```
