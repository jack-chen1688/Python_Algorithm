# 4. Median of Two Sorted Arrays

There are two sorted arrays nums1 and nums2 of size m and n respectively.

Find the median of the two sorted arrays. The overall run time complexity should be O(log (m+n)).

Example 1:
nums1 = [1, 3]
nums2 = [2]

The median is 2.0
Example 2:
nums1 = [1, 2]
nums2 = [3, 4]

The median is (2 + 3)/2 = 2.5

## Solution. O(log(min(M,N)))

```python
class Solution(object):
    def findMedianSortedArrays(self, nums1, nums2):
        """
        :type nums1: List[int]
        :type nums2: List[int]
        :rtype: float
        """
        def findMedian(nums):
            if nums == []:
                return None
            
            n = len(nums)
            if n & 1:
                return nums[n/2]
            else:
                return (nums[n/2] + nums[n/2 -1])/float(2)
        
        """
        Find out a number in nums1 and nums2 that is totally less than lt_num numbers in
        these two arrays
        """
        def findNum(lt_num, nums1, nums2):
            len1, len2 = len(nums1), len(nums2)
            lo, hi = 0, len1 - 1 
            
            # The loop is to find out the nu
            while lo <= hi: 
                mid = (lo + hi)/2
                #mid+1 to len1 -1, nums[mid] is less than or equal to lt_num1 numbers in nums1
                lt_num1 = len1 - 1 - mid
                lt_num2 = lt_num - lt_num1
                
                if lt_num2 < 0: 
                    lo = mid + 1
                    
                elif lt_num2 == 0:
                    if nums1[mid] >= nums2[-1]:
                        return nums1[mid]
                    else:
                        lo = mid + 1
                        
                else: #lt_num2 >0
                    if len2 < lt_num2 or nums1[mid] > nums2[-rem]:
                        # < lt_num2 case
                        hi = mid-1
                    else:
                        if (len2 == rem and nums1[mid] <= nums2[-rem]) \
                            or (len2 > rem and nums1[mid] >= nums2[-rem-1]):
                            return nums1[mid]
                        else:
                        # > lt_num
                            lo = mid + 1

            
            lt_num1 = len1 -1 - lo
            rem = lt_num - lt_num1 - 1
            #print lt_num1, rem
            return nums2[-rem-1]
        
        len1, len2 = len(nums1), len(nums2)
        
        if len1 > len2:
            len1, len2, nums1, nums2 = len2, len1, nums2, nums1
        
        if nums1 == []:
            return findMedian(nums2)
        
        total = len1 + len2
        lt_num = (total - 1)/2
        median_hi = findNum(lt_num, nums1, nums2)
        #print "median_hi", median_hi
        if total & 1 == 1:
            return median_hi
        else:
            median_lo = findNum(lt_num+1, nums1, nums2)
            return (median_hi + median_lo)/float(2)
```

## Solution 3. Enhancement to Find out the Lower Midian

```python
class Solution(object):
    def findMedianSortedArrays(self, nums1, nums2):
        """
        :type nums1: List[int]
        :type nums2: List[int]
        :rtype: float
        """
        def findMedian(nums):
            if nums == []:
                return None
            
            n = len(nums)
            if n & 1:
                return nums[n/2]
            else:
                return (nums[n/2] + nums[n/2 -1])/float(2)

        
        len1, len2 = len(nums1), len(nums2)
        
        if len1 > len2:
            len1, len2, nums1, nums2 = len2, len1, nums2, nums1
        
        if nums1 == []:
            return findMedian(nums2)

        total = len1 + len2
        lt_num = (total - 1)/2
        lo, hi = 0, len1 - 1 
        median_hi = None
        
        # The loop is to find out the nu
        while lo <= hi: 
            mid = (lo + hi)/2
            #mid+1 to len1 -1, nums[mid] is less than or equal to lt_num1 numbers in nums1
            lt_num1 = len1 - 1 - mid
            lt_num2 = lt_num - lt_num1
            
            if lt_num2 < 0: 
                lo = mid + 1

            elif lt_num2 == 0:
                if nums1[mid] >= nums2[-1]:
                    median_hi = nums1[mid]
                    break
                else:
                    lo = mid + 1

            else: #lt_num2 >0
                if len2 < lt_num2:
                    #or nums1[mid] > nums2[-rem]:
                    # < lt_num2 case
                    hi = mid-1
                elif len2 == lt_num2:
                    if nums1[mid] <= nums2[-len2]:
                        median_hi = nums1[mid]
                        break
                    else: # < lt_num2 case
                        hi = mid - 1
                else: #len2 > it_num2
                    if nums1[mid] < nums2[-lt_num2-1]:
                        # > lt_num2
                        lo = mid + 1
                    elif nums1[mid] > nums2[-lt_num2]:
                        # < lt_num2
                        hi = mid -1
                    else:
                        median_hi = nums1[mid]
                        break
            
            
        if median_hi == None:
            # Not found the number in nums1. nums1[lo] is the upper bound after the loop. Use nums1[lo] as base
            # to find out the number in nums2.
            lt_num1 = len1 - lo  # The number in nums2 is less than nums1[lo:len1].
            lt_num2 = lt_num - lt_num1
            median_hi = nums2[-lt_num2-1] 
            if total & 1 == 1:
                return median_hi
            else:
                if lo == 0:
                    median_lo = nums2[-lt_num2-2]
                elif lt_num2 + 2 > len2:
                    median_lo = nums1[lo - 1]
                else:
                    median_lo = max(nums2[-lt_num2-2], nums1[lo-1])
        else:
            if total & 1 == 1:
                return median_hi
            else:
                if mid == 0:
                    median_lo = nums2[-lt_num2 -1]
                elif lt_num2 + 1 > len2:
                    median_lo = nums1[mid-1]
                else:
                    median_lo = max(nums2[-lt_num2-1], nums1[mid-1])

        return (median_lo + median_hi)/2.0 


lst1, lst2 = [101], [100]
print Solution().findMedianSortedArrays(lst1,lst2)
```
