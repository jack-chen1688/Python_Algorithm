# 30. Substring with Concatenation of All Words

You are given a string, s, and a list of words, words, that are all of the same length. Find all starting indices of substring(s) in s that is a concatenation of each word in words exactly once and without any intervening characters.

For example, given:
s: "barfoothefoobarman"
words: ["foo", "bar"]

You should return the indices: [0,9].
(order does not matter).


## Solution 1. Time Exceeded

```python
class Solution(object):
    
    def findSubstring(self, s, words):
        """
        :type s: str
        :type words: List[str]
        :rtype: List[int]
        """
        word_cnt = len(words)
        word_len = len(words[0])
        str_len = len(s)
        
        dict = {}
        for word in words:
            if word not in dict.keys():
                dict[word] = 1
            else:
                dict[word] += 1
        
        result = []
        start_pos = 0
        pos = 0
        count = 0
        dict1 = dict.copy()
        
        while pos <= str_len - word_len:
            # get next token
            #print pos
            next_token = s[pos:pos+word_len]
            #print pos, next_token, round, dict
            if next_token in dict1.keys():
                
                if count == 0:
                   start_pos = pos
                    
                count += 1  
                dict1[next_token] -= 1
                
                if dict1[next_token] < 0:
                    # not matching, 
                    count = 0
                    pos = start_pos + 1
                    start_pos = pos
                    dict1 = dict.copy()
                else:
                    if count == word_cnt:
                        # find one set, restart from start_pos + 1
                        result.append(start_pos)
                        count = 0
                        pos = start_pos+1
                        dict1 = dict.copy()
                    else:
                        # find next token
                        pos += word_len
                        
            else:
                # token not matching
                if count != 0:
                    count = 0
                    pos = start_pos + 1
                    dict1 = dict.copy()
                else:
                    pos += 1

        return result

def main():
    s = "barfoothefoobarman"
    words = ["foo", "bar"]
    print Solution().findSubstring(s, words)
    
    s = "barfoofoobarman"
    words = ["foo", "bar"]
    print Solution().findSubstring(s, words)

    s = "bar1foo2foo1bar1man2"
    words = ["foo1", "bar1"]
    print Solution().findSubstring(s, words)
    
    s = "barfoofoobarthefoobarman"
    words = ["bar","foo","the"]
    print Solution().findSubstring(s, words)
    
if __name__ == "__main__":
    main()
            
                    
```

## Solution 3. Optimization Using O(n) Time

```python
class Solution(object):
    def findSubstring(self, s, words):
    
        # dict{word: number of repetition in words}
        dict={}
        for word in words:
            if word in dict.keys():
                dict[word] += 1
            else:
                dict[word] = 1
        
        word_cnt = len(words)
        word_len = len(words[0])
        str_len = len(s)
        
        result = []  
        
        for pos in range (word_len):
            start_pos = pos
            count = 0
            dict1 = dict.copy()
                      
            while pos + word_len <= str_len:
                next_token = s[pos:pos+word_len]
                if next_token in dict1.keys():
                    dict1[next_token] -= 1
                    
                    if count == 0:
                        start_pos = pos
                    
                    count += 1
                    if dict1[next_token] < 0:
                        # current token has repeated one more time. Search from start_pos 
                        # to find out the token that equals the current token. The token 
                        # after the found one can be the next start position.
                        found = 0
                        while start_pos < pos:
                            
                            if s[start_pos: start_pos + word_len] == next_token:
                                found = 1
                                
                            dict1[s[start_pos: start_pos + word_len]] += 1
                            start_pos += word_len
                            count -= 1
                            if found:
                                break
                    else:
                        # still matching
                        if count == word_cnt:  
                            # last token match, start again from start_pos+word_len
                            result.append(start_pos)
                            dict1[s[start_pos: start_pos + word_len]] += 1
                            start_pos += word_len
                            count -= 1 
                        # else:  continue with next token
                else:
                    # token not matching, start again from pos + word_len
                    count = 0
                    start_pos = pos
                    dict1 = dict.copy()
                    
                pos = pos + word_len
                
        return result

    
def main():
    s = "barfoothefoobarman"
    words = ["foo", "bar"]
    print Solution().findSubstring(s, words)
    
    s = "barfoofoobarman"
    words = ["foo", "bar"]
    print Solution().findSubstring(s, words)

    s = "bar1foo2foo1bar1man2"
    words = ["foo1", "bar1"]
    print Solution().findSubstring(s, words)
    
    s = "barfoofoobarthefoobarman"
    words = ["bar","foo","the"]
    print Solution().findSubstring(s, words)
    
if __name__ == "__main__":
    main()
```
